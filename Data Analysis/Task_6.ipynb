```python
#  Which country has won the most gold medals?
max_gold_trophies = df['Gold'].max()
df[df['Gold'] == max_gold_trophies]
```

```python
# How many countries won more than 10 gold medals?
df[df['Gold'] > 10]
```

```python
# Which country has the highest total number of medals?
df[df['Total'] == df['Total'].max()]
```

```python
# What is the average number of silver medals won by countries? and country who has won some  that much silver medal.
print("Average number of silver medals won by countries: ",df['Silver'].mean())
df[df['Silver'] == round(df['Silver'].mean())]
```

```python
# How many countries have won at least one medal (gold, silver, or bronze)?
df[(df['Gold'] >=1) | (df['Silver'] >= 1) | (df['Bronze'] >= 1)]['Country'].count()
```

```python
# What is the median number of bronze medals won by countries? and country who has won exactly that much bronze medal
print('median number of bronze medals won by countries: ',df['Bronze'].median())
df[df['Bronze'] == round(df['Bronze'].median())]
```

```python
#  Which country ranks lowest in terms of total medals won?
df[df['Total'] == df['Total'].min()]
```

```python
# Bar Plot of Top 10 Countries by Gold Medals

top10CountryByGoldMedals = df.sort_values(by='Gold', ascending=False).head(10)

plt.figure(figsize = (12, 6))
sns.barplot(x= 'Gold', y= 'Country', data = top10CountryByGoldMedals)
plt.title('Top 10 Countries by Gold Medals')
plt.xlabel('Number of Gold Medals')
plt.ylabel('Country')
plt.show()
```

```python
# Stacked Bar Plot of Medals for Top 10 Countries

top10CountryByMedals = df.sort_values(by='Total', ascending=False).head(10)

top10CountryByGoldMedals.set_index('Country')[['Gold', 'Silver', 'Bronze']].plot(kind='bar', stacked=True, color = ['gold', 'silver', '#cd7f32'], figsize = (10, 6))
plt.title('Medal Distribution for Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Number of Medals')
plt.show()
```

```python
# Pie Chart of Total Medals by Top 5 Countries

top5CountryByMedals = df.sort_values(by='Total', ascending=False).head(5)

plt.pie(top10CountryByMedals['Total'], labels = top10CountryByMedals['Country'])
plt.title('Total Medals Distribution among Top 5 Countries')
plt.show()
```

```python
# Histogram of Gold Medals

plt.figure(figsize = (10, 6))
sns.histplot(df['Gold'], bins = 20, kde = False, color = 'Gold')
plt.title('Distribution of Gold Medals')
plt.xlabel('Number of Gold Medals')
plt.ylabel('Number of Countries')
plt.show()
```

```python
# Boxplot of Medal Counts

plt.figure(figsize = (10, 6))
sns.boxplot(data = df[['Gold', 'Silver', 'Bronze']])
plt.title('Boxplot of Medal Counts')
plt.xlabel('Medal Type')
plt.ylabel('Number of Medals')
plt.show()
```

```python
# Line Plot of Cumulative Medals by Top 5 Countries

plt.figure(figsize=(10, 6))
sns.lineplot(data=top5CountryByMedals, x='Rank', y='Total', hue='Country', marker='o')
plt.title('Cumulative Medals by Top 5 Countries')
plt.xlabel('Rank')
plt.ylabel('Total Medals')
plt.show()
```

```python
# Bar Plot of the Average Number of Medals by Medal Type

avg_medals = df[['Gold', 'Silver', 'Bronze']].mean()

plt.figure(figsize=(8, 6))
sns.barplot(x=avg_medals.index, y=avg_medals.values, palette=['gold', 'silver', '#cd7f32'])
plt.title('Average Number of Medals by Type')
plt.xlabel('Medal Type')
plt.ylabel('Average Number of Medals')
plt.show()
```

```python
# Heatmap of Correlation Between Medal Types

plt.figure(figsize=(8, 6))
sns.heatmap(df[['Gold', 'Silver', 'Bronze', 'Total']].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Between Medal Types')
plt.show()
```
