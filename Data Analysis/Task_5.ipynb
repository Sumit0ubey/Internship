```python
# Are There Any Important Interactions Between Features?
# Pairplot for Interactions

sns.pairplot(df, vars=['cp', 'thalach', 'slope', 'oldpeak', 'exang'], hue='target', palette='coolwarm')
plt.suptitle('Pairplot of key Features with target', y=1.02)
plt.show()
```

```python
# How Do the Distributions of Features Vary for Different Classes?
# Violin Plots for Distribution Comparison

features = ['thalach', 'oldpeak', 'chol']
for feature in features:
    plt.figure(figsize=(10, 6))
    sns.violinplot(x='target', y=feature, data=df, palette='coolwarm')
    plt.title(f'Distribution of {feature} by Target Variable')
    plt.show()
```

```python

# Create a new feature and visualize
df1 = df
df1['chol_to_trestbps_ratio'] = df['chol'] / df['trestbps']

plt.figure(figsize=(10,6))
sns.scatterplot(x='chol_to_trestbps_ratio', y='target', hue='target', palette='coolwarm', data=df)
plt.title("Cholesterol to Resting BP Ratio vs Heart Disease")
plt.show()
```

```python
# Age Segmentation and Visualization

age_bins = [0, 40, 60, 100]
age_labels = ['<40', '40-60', '>60']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='thalach', y='chol', data=df, hue='target', style='age_group', palette='coolwarm')
plt.title('Max Heart Rate vs Cholesterol Across Age Groups')
plt.show()
```

```python
# Identifying Outliers

features = ['chol', 'trestbps', 'thalach', 'oldpeak']
for feature in features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='target', y=feature, data=df, palette='coolwarm')
    plt.title(f'Boxplot of {feature} by Target Variable')
    plt.show()
```
